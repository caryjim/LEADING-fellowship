Implemented the times series plot to see the peak in the University of Texas data

```python
!pip install statsmodels

from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
```

```python
utexas.sort_index(inplace=True)
```

```python
# Additive Decomposition
add_decomp = seasonal_decompose(utexas['clicks'], model='additive', extrapolate_trend='freq', period = 7) # for now I'm using 7 for weekly cycle 

# Multiplicative Decomposition
# multi_decomp = seasonal_decompose(data['clicks'], model='multiplicative', extrapolate_trend='freq', period = 7) # for now I'm using 7 for weekly cycle 

#For graphing 
trend = add_decomp.trend
seasonal = add_decomp.seasonal
residual = add_decomp.resid


# Plot gathered statistics
plt.figure(figsize=(12,8))
plt.subplot(411)
plt.plot(utexas['clicks'], label='Original', color="blue")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend', color="blue")
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality', color="blue")
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals', color="blue")
plt.legend(loc='best')
plt.tight_layout()

# summary statistics
print(residual.describe())

np.mean(residual)
```
